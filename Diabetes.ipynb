**DIABETES PREDICTION USING SVM**

**Importing Liabraries**

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler          # Upgrades data values within some range
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

**Data Analysis and Preprocessing**

```python
diabetes_data = pd.read_csv("/content/diabetes.csv")
diabetes_data
```

```python
diabetes_data.shape # 9 features with 768 training examples
```

```python
diabetes_data.describe()
```

```python
diabetes_data['Outcome'].value_counts()
```

0 = Non-diabetic

1 = Diabetic

```python
diabetes_data.groupby('Outcome').mean()
```

```python
X = diabetes_data.drop(columns='Outcome', axis=1)
Y = diabetes_data['Outcome']
X, Y
```

```python
# STANDARDIZING DATA
scaler = StandardScaler()
scaler.fit(X)
X_train_standarized = scaler.transform(X)
X = X_train_standarized
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=1, stratify=Y)
X_train, X_test, Y_train, Y_test
```

**Training Model**

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
```

**Mode Evaluation**

```python
# accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_accuracy = accuracy_score(X_train_prediction, Y_train)
training_accuracy
```
